1. ¿Existe una relación entre el nivel de educación y el tipo de crédito otorgado?
Hipótesis: Los individuos con educación superior tienden a recibir créditos más grandes que aquellos con educación secundaria o técnica.

Análisis y Visualización: Podemos usar un gráfico de caja (boxplot) para comparar los montos de crédito (AMT_CREDIT) entre diferentes niveles educativos (NAME_EDUCATION_TYPE).

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Cargar los datos desde la URL proporcionada, especificando el delimitador como ';'
url = "https://raw.githubusercontent.com/DavidGN2016GFS/DataScience/refs/heads/main/Riesgo_crediticio_limpio.csv"
data = pd.read_csv(url, delimiter=';')

# Eliminar espacios en los nombres de columnas
data.columns = data.columns.str.strip()

# Verificar los nombres de las columnas
print("Columnas en el DataFrame:", data.columns)

# Continuar con el análisis si el nombre de la columna es correcto
if 'NAME_EDUCATION_TYPE' in data.columns and 'AMT_CREDIT_SUM' in data.columns:
    # Calcular el promedio del monto de crédito por nivel educativo
    education_credit_avg = data.groupby('NAME_EDUCATION_TYPE')['AMT_CREDIT_SUM'].mean()

    # Preparación de los datos para el gráfico de radar
    categories = education_credit_avg.index.tolist()  # Obtener los niveles educativos
    values = education_credit_avg.values  # Obtener los valores promedio de crédito

    # Convertir valores en lista y agregar el primer valor al final para cerrar el gráfico
    values = np.append(values, values[0])
    angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False).tolist()
    angles += angles[:1]  # Añadir el primer ángulo al final para cerrar el gráfico

    # Crear el gráfico de radar
    fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
    ax.fill(angles, values, color='skyblue', alpha=0.4)
    ax.plot(angles, values, color='blue', linewidth=2)
    ax.set_yticklabels([])  # Eliminar etiquetas del eje radial

    # Añadir etiquetas para cada categoría
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(categories, fontsize=10)
    plt.title("Promedio de Monto de Crédito por Nivel Educativo", fontsize=15)
    plt.show()


```

Estadística Descriptiva
Cálculo de Medidas de Resumen: Calculamos medidas como el promedio, la mediana y el rango intercuartílico (IQR) del monto de crédito (AMT_CREDIT) para cada categoría de nivel educativo (NAME_EDUCATION_TYPE).

Promedio: Nos indica el monto medio de crédito dentro de cada nivel de educación, proporcionando una idea general de la tendencia.
Mediana: El valor central del monto de crédito en cada grupo educativo, menos afectado por valores extremos.
Rango Intercuartílico (IQR): Muestra la dispersión de los montos de crédito en el 50% medio de los datos y ayuda a identificar si existen diferencias significativas en la distribución entre los niveles educativos.
Comparación de las Medidas: Analizamos si el promedio y la mediana son consistentemente mayores en los niveles de educación superior que en los niveles secundarios o técnicos.

```python
import pandas as pd

# Cargar el dataset desde la URL con el delimitador correcto
url = "https://raw.githubusercontent.com/DavidGN2016GFS/DataScience/refs/heads/main/Riesgo_crediticio_limpio.csv"
data = pd.read_csv(url, sep=';')


# 1. Promedio (Media)
mean_credit = data.groupby('NAME_EDUCATION_TYPE')['AMT_CREDIT_SUM'].mean()

# 2. Mediana
median_credit = data.groupby('NAME_EDUCATION_TYPE')['AMT_CREDIT_SUM'].median()

# 3. Rango Intercuartílico (IQR)
q75 = data.groupby('NAME_EDUCATION_TYPE')['AMT_CREDIT_SUM'].quantile(0.75)
q25 = data.groupby('NAME_EDUCATION_TYPE')['AMT_CREDIT_SUM'].quantile(0.25)
iqr_credit = q75 - q25

# Mostrar resultados
print("Promedio del Monto de Crédito por Nivel Educativo:")
print(mean_credit)
print("\nMediana del Monto de Crédito por Nivel Educativo:")
print(median_credit)
print("\nRango Intercuartílico (IQR) del Monto de Crédito por Nivel Educativo:")
print(iqr_credit)


```

Conclusión

Promedio del Monto de Crédito:

Los niveles educativos con los montos promedio de crédito más altos son "Incomplete higher" (380,937) y "Higher education" (360,388), lo cual indica que las personas con algún grado de educación superior tienden a obtener montos de crédito más altos en promedio.
Los niveles educativos con montos de crédito promedio más bajos son "Academic degree" (266,437) y "Lower secondary" (309,260), lo cual es algo contradictorio en el caso de "Academic degree" ya que se esperaría que aquellos con títulos académicos tengan acceso a mayores montos.
Mediana del Monto de Crédito:

La mediana, que representa el valor central, muestra que "Academic degree" tiene una mediana más alta (180,000) en comparación con otros niveles, lo que sugiere que aunque el promedio no sea el más alto, muchos de los créditos otorgados están en niveles elevados.
Los otros niveles ("Higher education", "Incomplete higher", "Lower secondary", y "Secondary / secondary special") tienen medianas alrededor de 117,000 a 126,000, indicando que la mayoría de los créditos en estos grupos están en rangos similares.
Rango Intercuartílico (IQR):

"Academic degree" tiene el IQR más amplio (431,230.5), lo que indica una mayor variabilidad en los montos de crédito otorgados para este grupo.
"Higher education" y "Incomplete higher" también tienen rangos intercuartílicos altos (257,981.94 y 294,944.625 respectivamente), lo cual implica una diversidad en los montos otorgados, aunque menor que en el caso de "Academic degree".
Los niveles "Lower secondary" y "Secondary / secondary special" muestran una menor variabilidad en los montos de crédito, con un IQR menor en comparación con los niveles educativos superiores.
Conclusión
Aunque el análisis no muestra una relación perfectamente lineal entre el nivel de educación y los montos de crédito, sí podemos observar una tendencia: los individuos con algún grado de educación superior ("Higher education" y "Incomplete higher") tienden a recibir mayores montos de crédito en promedio. Sin embargo, el grupo con "Academic degree" presenta montos promedio menores, lo cual podría ser atribuido a características específicas de la muestra o a criterios de riesgo aplicados por la institución crediticia.

En resumen, existe una relación entre el nivel educativo y el monto de crédito otorgado, donde niveles educativos más altos suelen estar asociados con créditos mayores, aunque con algunas excepciones en el caso de aquellos con títulos académicos.

2. ¿El número de hijos influye en la cantidad total de ingresos?
Hipótesis: Las personas con más hijos tienen mayores ingresos debido a sus responsabilidades familiares.

Análisis y Visualización: Para investigar esta hipótesis, podemos hacer un gráfico de dispersión (scatterplot) del ingreso total (AMT_INCOME_TOTAL) en función de la cantidad de hijos (CNT_CHILDREN) y agregar una línea de tendencia.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset desde el archivo CSV
url = "https://raw.githubusercontent.com/DavidGN2016GFS/DataScience/refs/heads/main/Riesgo_crediticio_limpio.csv"
data = pd.read_csv(url, sep=';')

# Análisis descriptivo de AMT_INCOME_TOTAL en función de CNT_CHILDREN
summary = data.groupby('CNT_CHILDREN')['AMT_INCOME_TOTAL'].describe()
print("Resumen descriptivo del ingreso total por número de hijos:")
print(summary)

# Cálculo adicional: Rango Intercuartílico (IQR) de AMT_INCOME_TOTAL por número de hijos
iqr_income = data.groupby('CNT_CHILDREN')['AMT_INCOME_TOTAL'].apply(lambda x: x.quantile(0.75) - x.quantile(0.25))
print("\nRango Intercuartílico del ingreso total por número de hijos:")
print(iqr_income)

# Gráfico de dispersión del ingreso total en función de la cantidad de hijos, con una línea de tendencia
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.scatterplot(x='CNT_CHILDREN', y=data['AMT_INCOME_TOTAL'] / 1000, data=data, alpha=0.5)
sns.regplot(x='CNT_CHILDREN', y=data['AMT_INCOME_TOTAL'] / 1000, data=data, scatter=False, color='red', line_kws={"linewidth": 2})
plt.title("Ingreso Total en función del Número de Hijos")
plt.xlabel("Número de Hijos")
plt.ylabel("Ingreso Total (en miles de dólares)")

# Gráfico de cajas para visualizar la distribución de ingresos en cada grupo de hijos
plt.subplot(1, 2, 2)
sns.boxplot(x='CNT_CHILDREN', y=data['AMT_INCOME_TOTAL'] / 1000, data=data, palette="Blues")
plt.title("Distribución de Ingreso Total por Número de Hijos")
plt.xlabel("Número de Hijos")
plt.ylabel("Ingreso Total (en miles de dólares)")

# Mostrar ambos gráficos
plt.tight_layout()
plt.show()

```

Análisis de los datos
Media del Ingreso Total:

Las personas sin hijos tienen un ingreso promedio de 168,159, mientras que aquellas con uno o más hijos presentan ingresos ligeramente más altos (por ejemplo, 173,141 con un hijo y 172,940 con dos hijos).
Las personas con tres o cuatro hijos tienen ingresos promedio aún mayores (189,897 y 194,538, respectivamente).
Sin embargo, para las personas con cinco hijos, el ingreso promedio es menor (159,750), lo cual rompe un poco la tendencia.
Mediana del Ingreso Total:

La mediana del ingreso muestra una tendencia similar a la de la media, con ligeros incrementos a medida que aumenta el número de hijos, aunque esta variación no es significativa. Esto indica que, en general, el ingreso no cambia drásticamente en función del número de hijos.
Rango Intercuartílico (IQR):

El IQR, que mide la dispersión en los ingresos, tiende a ser mayor para aquellos con más hijos (especialmente para quienes tienen entre 3 y 4 hijos), lo que sugiere una mayor variabilidad en los ingresos de estos grupos.
Para personas con cinco hijos, el IQR es menor (65,250), lo cual indica que los ingresos son menos variables en este grupo.
Observación de la Visualización
El gráfico de dispersión muestra que la mayoría de los puntos están agrupados en un rango bajo de ingresos, sin importar el número de hijos, y la línea de tendencia es prácticamente horizontal, lo cual sugiere que no hay una correlación significativa entre el número de hijos y el ingreso total.

Conclusión
Aunque las personas con más hijos tienden a tener ingresos promedio ligeramente más altos, la diferencia en ingresos según el número de hijos no es considerable y, en algunos casos (como para personas con cinco hijos), el ingreso promedio es incluso menor. La línea de tendencia casi horizontal en el gráfico de dispersión también indica una falta de relación fuerte entre estas dos variables.

En resumen, no hay suficiente evidencia para afirmar que el número de hijos influya significativamente en el ingreso total

4. ¿El estado laboral actual (DAYS_EMPLOYED) influye en el tipo de crédito activo y la cantidad de crédito otorgada?
Hipótesis: Las personas que han estado empleadas durante más tiempo reciben créditos más grandes y con mayor frecuencia que las personas con menos tiempo de empleo.

Análisis y Visualización: Podemos hacer un scatterplot para observar la relación entre los días empleados (DAYS_EMPLOYED) y el monto de crédito (AMT_CREDIT), y añadir una línea de regresión para observar la tendencia.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

# Cargar el dataset desde el archivo CSV
url = "https://raw.githubusercontent.com/DavidGN2016GFS/DataScience/refs/heads/main/Riesgo_crediticio_limpio.csv"
data = pd.read_csv(url, sep=';')

# Eliminar el valor erróneo 365243 y convertir los valores negativos de DAYS_EMPLOYED en positivos
data = data[data['DAYS_EMPLOYED'] != 365243]  # Eliminar el valor erróneo 365243
data['DAYS_EMPLOYED'] = data['DAYS_EMPLOYED'].abs()  # Convertir los valores negativos en positivos

# Resumen descriptivo de días empleados y monto de crédito
summary_employment_credit = data[['DAYS_EMPLOYED', 'AMT_CREDIT_SUM']].describe()
print("Resumen descriptivo de Días Empleados y Monto de Crédito:")
print(summary_employment_credit)

# Resumen por estado de crédito
credit_status_summary = data.groupby('CREDIT_ACTIVE')['AMT_CREDIT_SUM'].describe()
print("\nResumen del Monto de Crédito por Estado del Crédito:")
print(credit_status_summary)

# Distribución de días empleados por estado del crédito
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='DAYS_EMPLOYED', hue='CREDIT_ACTIVE', multiple="stack", kde=True)
plt.title("Distribución de Días Empleados por Estado del Crédito")
plt.xlabel("Días Empleado")
plt.ylabel("Frecuencia")
plt.show()

# Gráfico de dispersión y línea de tendencia entre días empleados y monto de crédito
plt.figure(figsize=(12, 8))
sns.scatterplot(x='DAYS_EMPLOYED', y=data['AMT_CREDIT_SUM'] / 1000, hue='CREDIT_ACTIVE', data=data, alpha=0.6)
sns.regplot(x='DAYS_EMPLOYED', y=data['AMT_CREDIT_SUM'] / 1000, data=data, scatter=False, color='blue', line_kws={"linewidth": 2})

# Títulos y etiquetas
plt.title("Relación entre Tiempo de Empleo y Monto de Crédito")
plt.xlabel("Días Empleado")
plt.ylabel("Monto de Crédito (en miles)")
plt.legend(title='Estado del Crédito')
plt.show()

# Prueba de correlación entre Días Empleado y Monto de Crédito
corr_coef, p_value = stats.pearsonr(data['DAYS_EMPLOYED'], data['AMT_CREDIT_SUM'])
print(f"\nCoeficiente de correlación entre Días Empleado y Monto de Crédito: {corr_coef:.3f}")
print(f"Valor p: {p_value:.3f}")
```

5. ¿Qué relación existe entre la antigüedad del empleo y el valor de los bienes adquiridos?
Hipótesis: A mayor antigüedad en el empleo, mayor es el valor de los bienes que se pueden adquirir mediante crédito.

Análisis y Visualización: Usaremos un scatterplot para observar la relación entre los días empleados (DAYS_EMPLOYED) y el valor de los bienes (AMT_GOODS_PRICE), con una línea de tendencia.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset desde el archivo CSV usando el enlace
url = "https://raw.githubusercontent.com/DavidGN2016GFS/DataScience/refs/heads/main/Riesgo_crediticio_limpio.csv"
data = pd.read_csv(url, sep=';')

# Filtrar valores dañados en DAYS_EMPLOYED
data = data[data['DAYS_EMPLOYED'] != 365243]  # Eliminar el valor 365243

# Convertir los valores negativos de DAYS_EMPLOYED a positivos
data['DAYS_EMPLOYED'] = data['DAYS_EMPLOYED'].abs()

# Verificar si hay valores nulos en las columnas relevantes
print("Valores nulos en las columnas relevantes:")
print(data[['DAYS_EMPLOYED', 'AMT_GOODS_PRICE']].isnull().sum())

# Filtrar clientes activos (donde el valor en DAYS_EMPLOYED es razonable)
data_active = data[data['DAYS_EMPLOYED'] < 365243]

# Eliminar valores nulos de los datos activos para las columnas necesarias
data_active = data_active.dropna(subset=['DAYS_EMPLOYED', 'AMT_GOODS_PRICE'])

# Asegurarnos de que las columnas no tengan valores no numéricos
data_active = data_active[pd.to_numeric(data_active['DAYS_EMPLOYED'], errors='coerce').notnull()]
data_active = data_active[pd.to_numeric(data_active['AMT_GOODS_PRICE'], errors='coerce').notnull()]

# Comprobar la cantidad de registros después de la limpieza
print(f"Cantidad de registros después de la limpieza: {data_active.shape[0]}")

# Asegurarnos de que los valores de 'DAYS_EMPLOYED' y 'AMT_GOODS_PRICE' están en rangos razonables
print("Estadísticas descriptivas de los datos activos:")
print(data_active[['DAYS_EMPLOYED', 'AMT_GOODS_PRICE']].describe())

# Crear el scatter plot para clientes activos
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data_active, x='DAYS_EMPLOYED', y='AMT_GOODS_PRICE', color='salmon')

# Añadir una línea de tendencia (regresión lineal)
sns.regplot(data=data_active, x='DAYS_EMPLOYED', y='AMT_GOODS_PRICE', scatter=False, color='blue', line_kws={"color": "blue", "lw": 2})

plt.title('Relación entre Antigüedad en el Empleo y Valor de los Bienes (Clientes Activos)')
plt.xlabel('Días Empleado')
plt.ylabel('Valor de los Bienes')
plt.show()


```
